```python
from enum import auto, IntFlag
from typing import Any

from pydantic import (
    BaseModel,
    EmailStr,
    Field,
    SecretStr,
    ValidationError,
)
```

Aqui é feita apenas a importação de algumas bibliotecas e ferramentas que serão usadas durante o código

```python
class Role(IntFlag):
    Author = auto()
    Editor = auto()
    Developer = auto()
    Admin = Author | Editor | Developer


class User(BaseModel):
    name: str = Field(examples=["Arjan"])
    email: EmailStr = Field(
        examples=["example@arjancodes.com"],
        description="The email address of the user",
        frozen=True,
    )
    password: SecretStr = Field(
        examples=["Password123"], description="The password of the user"
    )
    role: Role = Field(default=None, description="The role of the user")
```

Aqui são criadas as classe Role(Função) e User, que serão usadas para os testes. Detalhe interessante é que na classe Role é usado o "auto" combinado com o "intFlag", tal combinação permite que as funções sejam combinadas, como a função "Admin". Isso é possível pois o intFlag trabalha com bits, logo os valores são atribuidos às funções da forma: 100(4), 010(2), 001(1), e combinando-as da seguinte forma: 001(1) + 010(2) + 100(4) = 111(7). Na classe User, é usado validadores do pydantic, como EmailStr e SecretStr.

```python
def validate(data: dict[str, Any]) -> None:
    try:
        user = User.model_validate(data)
        print(user)
    except ValidationError as e:
        print("User is invalid")
        for error in e.errors():
            print(error)


def main() -> None:
    good_data = {
        "name": "Arjan",
        "email": "example@arjancodes.com",
        "password": "Password123",
    }
    bad_data = {"email": "<bad data>", "password": "<bad data>"}

    validate(good_data)
    validate(bad_data)


if __name__ == "__main__":
    main()
```

Aqui é feita função validate, que vai ser responsável por validar os dados, e em caso de dados incoerentes, ela retorna uma exceção especificando o erro que ocorreu. E na main é possível ver dois exemplos, um com dados corretos (good_data) e o outro com os dados errados (bad_data)
