Aqui é a parte mais importante do código, onde é de fato definido o modelo que vamos usar, onde criamos manualmente o nosso menu de items que vai ser usado como banco de dados para a aplicação e, támbem, onde organizamos como será o vetor que será enviado para a base de dados. Primeiro é criado uma lista vazia para armazenar os objetos que serão enviados para o qdrant, depois é feito um loop para juntar o nome do item com sua descrição, visando dar mais contexto ao modelo de IA. Por conseguinte, é feito outro loop para adicionar um id a cada item e organizar o formato dos dados, criando os campos "item_name", "description", "price" e "category", e mandando os respectivos textos para cada um deles e, no fim, adicionamos o objeto na lista vazia que foi criada anteriormente.

```python
# generate query embedding
query_text = "vegetarian dishes"
query_vector = next(iter(model.embed(query_text)))

# search for similar menu items
results = client.query_points(
    collection_name="items",
    query=query_vector,
    with_payload=True,
    limit=5
)

# print results
for result in results.points:
    print(f"Item: {result.payload.get('item_name', 'N/A')}")
    print(f"Score: {result.score}")
    print(f"Description: {result.payload['description'][:150]}...")
    print(f"Price: {result.payload.get('price', 'N/A')}")
    print("---")
```
